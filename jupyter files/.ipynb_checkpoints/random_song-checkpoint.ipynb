Your package contains 3 different methods:
 method 1 returns a random new song 
 method 2 returns the highest rated song (you choose your own rating mechanism)
 method 3 returns a song based upon an inserted search (key) word
Per book you need to show the following attributes:
 Title
 Author
 2 different scores
 Cover of the album

```python
import requests
from PIL import Image
from bs4 import BeautifulSoup
import random
import re
```

```python
random_song_url = requests.get("http://www.songlyrics.com/r-and-b-lyrics.php")
```

```python
#make it beautiful, remove the trash
entire_table = BeautifulSoup(random_song_url.text, features="html.parser").find('tbody')
```

```python
#print(entire_table)
```

```python
items = []
for el in entire_table.findAll('tr'):
        song_title = el.find('h3').text
        artist_name = el.find('span').text
        
        find_image = el.find('img', attrs={"class": "tiny-img"})
        imgfile = find_image.get("src")
        im = Image.open(requests.get(imgfile, stream=True).raw)
        items.append((song_title,artist_name,im))
#         print(song_title,artist_name)        
```

```python
# print(len(items))
```

```python
rand_item = random.choice(items) 

img = rand_item[2]
name = rand_item[1]
song = rand_item[0]
print("Title: ",song)
print("Artist: " , name)
img
```

```python
song = re.sub(" ", "-", song)
artist_name = re.sub(" ","-",name)

res = requests.get(f'https://genius.com/{artist_name}-{song}-lyrics')
#print(res)
art_text = BeautifulSoup(res.text, 'html.parser')
spans = art_text.find_all('span', {'class': 'LabelWithIcon__Label-sc-1ri57wg-1 kMItKF'})
#print(spans)
check = 0
for span in spans:
    if(check == 1):
        views = span.text
        check = 0
    else:
        check = 1
```

```python
# get likes data
rest = requests.get(f'http://www.songlyrics.com/{artist_name}/{song}/')
art_te = BeautifulSoup(rest.text, 'html.parser')
span = art_te.find('span', {'id': 'fbbox-total'})
likes = re.sub(r'(\n?|\t?)','',span.text)
```

```python
print("Title: ",song)
print("Artist: " , name)
print('Views: ',views,' and Likes: ',likes)
```
